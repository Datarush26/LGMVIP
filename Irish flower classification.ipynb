```python
data['Species'].unique()
```

## 5.Spiliting the dataset

```python
from sklearn.model_selection import train_test_split
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.2)
```

## 6.Scale the data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

## 7.Model Creation

# 7.1 Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
pred=lr.predict(x_test)
```

```python
print('Acuuracy is :',(accuracy_score(y_test,pred)*100))
```

#### 7.1.1 Classification Report

```python
print(classification_report(y_test,pred))
```

#### 7.1.2 confusion matrix

```python
print(confusion_matrix(y_test,pred))
```

# 7.2 KNN classification

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
pred2=knn.predict(x_test)
print("Accuracy is :",accuracy_score(y_test,pred2)*100)
```

#### 7.2.1 Classification Report

```python
print(classification_report(y_test,pred2))
```

#### 7.2.2 confusion matrix

```python
print(confusion_matrix(y_test,pred2))
```

# 7.3 Naive bayes

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
pred3=nb.predict(x_test)
print("accuracy score of Naive bayes:",accuracy_score(y_test,pred3)*100)
```

#### 7.3.1 Classification Report

```python
print(classification_report(y_test,pred3))
```

#### 7.3.2 confusion matrix

```python
print(confusion_matrix(y_test,pred3))
```

# 7.4 Random Forest Classifier

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=40,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
r=RandomForestClassifier(n_estimators=20)
```

```python
r.fit(x_train,y_train)
pred4=r.predict(x_test)
print("The accuracy score of Randomforest is :",accuracy_score(y_test,pred4)*100)
```

#### 7.4.1 Classification Report

```python
print(classification_report(y_test,pred4))
```

#### 7.4.2 confusion matrix

```python
print(confusion_matrix(y_test,pred4))
```
